Then, we're going to use the term "program" to find synsets like so:

```python
syns = wordnet.synsets("program")
```

An example of a synset:

```python
print(syns[0].name())
```

Just the word:

```python
print(syns[0].lemmas()[0].name())
```

Definition of that first synset:

```python
print(syns[0].definition())
```

Examples of the word in use:

```python
print(syns[0].examples())
```

Next, how might we discern synonyms and antonyms to a word? The lemmas will be synonyms, and then you can use .antonyms to find the antonyms to the lemmas. As such, we can populate some lists like:

```python
synonyms = []
antonyms = []

for syn in wordnet.synsets("good"):
    for l in syn.lemmas():
        synonyms.append(l.name())
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())

print(set(synonyms))
print(set(antonyms))
```

Next, we can also easily use WordNet to compare the similarity of two words and their tenses, by incorporating the 'Wu and Palmer method' for semantic related-ness. Let's compare the noun of "ship" and "boat:"

```python
w1 = wordnet.synset('ship.n.01')
w2 = wordnet.synset('boat.n.01')
print(w1.wup_similarity(w2))
```

Now 'boat' and 'car'

```python
w1 = wordnet.synset('ship.n.01')
w2 = wordnet.synset('car.n.01')
print(w1.wup_similarity(w2))
```
